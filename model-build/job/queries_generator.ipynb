## Importing Basic Packages

```python
import pandas as pd
import random
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

## Importing Data

```python
branches=list(pd.read_csv('branches.csv'))
```

```python
states=list(pd.read_csv('states.csv'))
```

## Considering possible queries to create training and test datasets

### Query data generation and feature extraction

```python
yui=[]
for i in range(50):
    x=random.randint(0,5)
    yui.append(branches[x])
```

```python
tyu=[]
for i in range(50):
    x=random.randint(0,21)
    tyu.append(states[x])
```

```python
query=pd.DataFrame(list(zip(yui,tyu)),columns =['BRANCH','LOC_STATE'])
```

```python
with open('state_mapper.txt', 'rb') as handle:
    data = handle.read()
n = pickle.loads(data)
```

```python
regionq=[]
for v in query['LOC_STATE']:
    for w in n:
        if v in n[w]:
            regionq.append(w)
            break
```

```python
query['LOC_REGION']=regionq
```

```python
query.to_csv('queries.csv')
```

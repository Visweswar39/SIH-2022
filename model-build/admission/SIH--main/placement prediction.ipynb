```python
from sklearn.linear_model import LinearRegression, Ridge, RidgeCV, LassoCV, LassoLarsCV, ElasticNet, Lasso
from sklearn.preprocessing import scale, MinMaxScaler
import pickle
from sklearn.model_selection import cross_val_score
from sklearn import svm
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor
from sklearn.neural_network import MLPRegressor
from sklearn import linear_model
```

```python
#svm
svm_clf = svm.SVR()
svm_clf.fit(X,y)
ysvm_pred=svm_clf.predict(X1)
print('mae=',mean_absolute_error(y1,ysvm_pred))
print('mse=',mean_squared_error(y1,ysvm_pred))
print('r2_score=',r2_score(y1,ysvm_pred))
```

```python
#random forests
rand_clf=RandomForestRegressor(n_estimators = 1000,max_features='sqrt')
rand_clf.fit(X,y)
yrand_pred=rand_clf.predict(X1)
print('mae=',mean_absolute_error(y1,yrand_pred))
print('mse=',mean_squared_error(y1,yrand_pred))
print('r2_score=',r2_score(y1,yrand_pred))
```

```python
#lasso
lasso_clf = linear_model.Lasso(alpha=0.001, selection="random", random_state=1)
lasso_clf.fit(X,y)
ylasso_pred=lasso_clf.predict(X1)
print('mae=',mean_absolute_error(y1,ylasso_pred))
print('mse=',mean_squared_error(y1,ylasso_pred))
print('r2_score=',r2_score(y1,ylasso_pred))
```

```python
ridge_clf = Ridge(alpha=0.1)
ridge_clf.fit(X,y)
yridge_pred=ridge_clf.predict(X1)
print('mae=',mean_absolute_error(y1,yridge_pred))
print('mse=',mean_squared_error(y1,yridge_pred))
print('r2_score=',r2_score(y1,yridge_pred))
```
